```python
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score
from joblib import dump

#Import validity csv file, encoding 'E_GDP_1' as integers and deleting the column 'O_daval'
MG_val = pd.read_csv('MGdataset_P2_validation.csv', decimal =".", thousands = ",")
MG_val['E_GDP_1'] = pd.to_numeric(MGdata_P2['E_GDP_1'], downcast = 'integer')
del MG_val['O_daval']

MG_val["R_bids"]=MG_val["R_bids"].fillna(3.37)

#Encoding data that is non-numerical
MG_val['O_finstr'] = MG_val['O_finstr'].replace('IPA', 1)
MG_val['O_finstr'] = MG_val['O_finstr'].replace('ENI', 2)
MG_val['O_finstr'] = MG_val['O_finstr'].replace('EULEX', 3)

MG_val['O_countryp_code'] = MG_val['O_countryp_code'].replace('AL', 1)
MG_val['O_countryp_code'] = MG_val['O_countryp_code'].replace('BA', 2)
MG_val['O_countryp_code'] = MG_val['O_countryp_code'].replace('DZ', 3)
MG_val['O_countryp_code'] = MG_val['O_countryp_code'].replace('EG', 4)
MG_val['O_countryp_code'] = MG_val['O_countryp_code'].replace('JO', 5)
MG_val['O_countryp_code'] = MG_val['O_countryp_code'].replace('LB', 6)
MG_val['O_countryp_code'] = MG_val['O_countryp_code'].replace('ME', 7)
MG_val['O_countryp_code'] = MG_val['O_countryp_code'].replace('MK', 8)
MG_val['O_countryp_code'] = MG_val['O_countryp_code'].replace('MO', 9)
MG_val['O_countryp_code'] = MG_val['O_countryp_code'].replace('SB', 10)
MG_val['O_countryp_code'] = MG_val['O_countryp_code'].replace('TK', 11)
MG_val['O_countryp_code'] = MG_val['O_countryp_code'].replace('TN', 12)
MG_val['O_countryp_code'] = MG_val['O_countryp_code'].replace('XK', 13)

#Preprocessing: Selecting necessary columns, then splitting data into training and testing sets
MG_val = MG_val [['O_aval','O_finstr','R_bids','O_catpro','E_GDP_1','E_distance', 'O_dlots']]
MGdata_P2 = MGdata_P2[['O_aval','O_finstr','R_bids','O_catpro','E_GDP_1','E_distance', 'O_dlots'  ,'viable']]
X = MGdata_P2.drop('viable', axis=1)
y = MGdata_P2['viable']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

#Scaling the data prior to training model
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
viable_scaled =  scaler.transform(MG_val)
```

```python
from sklearn.metrics._plot.confusion_matrix import confusion_matrix
#Training a model with RandomForestClassifier and scoring accuracy
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
model = RandomForestClassifier(n_estimators=300, max_depth=8)
model.fit(X_train_scaled, y_train)
print('Feature Importance:', model.feature_importances_)

#Show statistics and confusion matrix
print('Training Accuracy:',accuracy_score(y_train, model.predict(X_train_scaled)))
print('Test Accuracy:',accuracy_score(y_test, model.predict(X_test_scaled)))
print('Viability Prediction:', model.predict(viable_scaled))
print('F1 score:',f1_score(y_test, model.predict(X_test_scaled)))
cm = confusion_matrix(y_test, model.predict(X_test_scaled))
sns.heatmap(cm, annot = True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
#Training a model with KNeighboursClassifier and scoring accuracy
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=4, p=1)
model.fit(X_train_scaled, y_train)

#Show statistics and confusion matrix
print('Training Accuracy:',accuracy_score(y_train, model.predict(X_train_scaled)))
print('Test Accuracy:',accuracy_score(y_test, model.predict(X_test_scaled)))
print('Viability Prediction:', model.predict(viable_scaled))
print('F1 score:',f1_score(y_test, model.predict(X_test_scaled)))
cm = confusion_matrix(y_test, model.predict(X_test_scaled))
sns.heatmap(cm, annot = True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
#Training a model with HistGradientBoostingClassifier and scoring accuracy
from sklearn.ensemble import HistGradientBoostingClassifier
model = HistGradientBoostingClassifier(max_depth=3)
model.fit(X_train_scaled, y_train)

#Show statistics and confusion matrix. This was our most accurate model, downloaded it as a .joblib file
print('Training Accuracy:',accuracy_score(y_train, model.predict(X_train_scaled)))
print('Test Accuracy:',accuracy_score(y_test, model.predict(X_test_scaled)))
print('Viability Prediction:', model.predict(viable_scaled))
print('F1 score:',f1_score(y_test, model.predict(X_test_scaled)))
cm = confusion_matrix(y_test, model.predict(X_test_scaled))
sns.heatmap(cm, annot = True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
dump(model,"A2model.joblib")
```

```python
#Training a model with AdaBoostClassifier and scoring accuracy
from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier(n_estimators=50)
model.fit(X_train_scaled, y_train)
print('Feature Importance:', model.feature_importances_)

#Show statistics and confusion matrix
print('Training Accuracy:',accuracy_score(y_train, model.predict(X_train_scaled)))
print('Test Accuracy:',accuracy_score(y_test, model.predict(X_test_scaled)))
print('Viability Prediction:', model.predict(viable_scaled))
print('F1 score:',f1_score(y_test, model.predict(X_test_scaled)))
cm = confusion_matrix(y_test, model.predict(X_test_scaled))
sns.heatmap(cm, annot = True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
